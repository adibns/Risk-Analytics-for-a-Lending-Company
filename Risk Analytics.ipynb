```python
# Drop columns with more than 50% null values
loan_df=loan_df.loc[:,round(loan_df.isnull().sum()/len(loan_df)*100,2)<50]
loan_df.shape
```

```python
#Remove columns with only one unique values
loan_df= loan_df.loc[:,loan_df.nunique()!=1]
loan_df.shape
```

```python
# Dropping column which we don't need for analysis
drop_columns=['id','member_id','funded_amnt_inv','emp_title','url','total_acc','desc','title','zip_code','issue_d','last_credit_pull_d','last_pymnt_d','total_rec_int','total_rec_late_fee','recoveries','collection_recovery_fee','last_pymnt_amnt','out_prncp','out_prncp_inv','installment','total_rec_prncp','total_pymnt_inv','earliest_cr_line']
loan_df=loan_df.drop(drop_columns,axis=1)
```

```python
# removing 'current' from loan status as it is not required
loan_df=loan_df[loan_df.loan_status !='Current']
loan_df=loan_df.loc[:,loan_df.nunique()!=1]
```

#### Clearly we can see too many outliers in annual income column and outliers in loan amount, funded amount and revolving balance columns. So we have to remove the outliers

```python
# Process of removing outliers by taking above 4 columns into new dataframe and then removing the outliers
df_outlier=loan_df[['loan_amnt','funded_amnt','annual_inc','revol_bal']]
df_outlier.head()
```

```python
# removing top and least 5% outliers
minim = 0.05
maxm = 0.95
df_outlier_sl = df_outlier.quantile([minim, maxm])
print(df_outlier_sl)
df_outlier = df_outlier.apply(lambda x: x[(x>df_outlier_sl.loc[minim,x.name]) & (x < df_outlier_sl.loc[maxm,x.name])], axis=0)
df_outlier_sl.head()
```

```python
# Changing the data from df_outliers to loan_df
loan_df[['loan_amnt','funded_amnt','annual_inc','revol_bal']] = df_outlier[['loan_amnt','funded_amnt','annual_inc','revol_bal']]
loan_df.head()
```

```python
# dropping na's
df = loan_df
df.dropna(inplace=True)
```
